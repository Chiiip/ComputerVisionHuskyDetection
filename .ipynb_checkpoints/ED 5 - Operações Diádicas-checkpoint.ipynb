## Carregando o vídeo

```python
from machinevisiontoolbox import VideoFile
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import math
import time
from IPython.display import clear_output

movie = VideoFile('Video2_husky.mp4', grey=True)
plt.gray()
```

## Iterando os frames dos vídeos e calculando a diferença entre eles

```python
previous_frame = []
i = 0
for frame in movie:
    if (i != 0):
        plt.subplots(figsize=(10, 8))
        diff = frame.to_float() - previous_frame
        imgplot = plt.imshow(np.abs(diff))
        plt.show()
        time.sleep(0.005)
        clear_output(wait=True)
    i = 1
    previous_frame = frame.to_float()
```
